On-Demand Software

software distribution model in which services(the backend) are hosted by a cloud service provider. These services are available to end-users over the internet(they access by frontend which is also hosted or an app) so, the end-users do not need to install any software on their devices to access these service

# Advantages  
SaaS is easy to buy  (monthly) in comparison of one time purchase  
The initial set-up cost for SaaS is typically less than the enterprise software

SaaS services are offered as a one-to-many model means a single instance of the application is shared by multiple users.
