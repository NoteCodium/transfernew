What is the Waterfall Model?
The Waterfall model is a sequential and linear approach to software development. Each phase must be completed before moving to the next, with little room for iteration.

Phases of the Waterfall Model:
Requirement Analysis: Gather and document all requirements.
System Design: Create detailed designs for architecture and modules.
Implementation: Write the code based on the design.
Testing: Verify that the software meets requirements.
Deployment: Release the final product to the customer.
Maintenance: Fix bugs and make minor updates.

Key Characteristics:
Rigid and structured process.
Minimal customer involvement after the requirements phase.
Changes are difficult and costly to implement once the project begins.
