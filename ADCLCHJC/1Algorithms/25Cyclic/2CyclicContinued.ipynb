```c++11
//generic method learned at lesson one
vector<int> sort(vector<int> &v) {
    int n=v.size();
    for(int i=0;i<n;i++){
      if(v[i]!=i+1){
        int idx=v[i]-1;
        int val=v[i];
        while(idx!=i){
          int tmp=v[idx];
          v[idx]=val;
          idx=tmp-1;
          val=tmp;
        }
        v[i]=val;  
      }
    }
    return v;
  }
```
