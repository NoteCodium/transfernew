```python
ll fff(ll a, ll b, ll c){
    return (a-b)*(a-b) + (b-c)*(b-c) + (c-a)*(c-a);
}


void ffff(ll &ans, vl &l, vl &m, vl &h){
    for(ll x: m){
        vl tmp1,tmp2;   
        auto it1=lower_bound(it(l),x);
        if(it1!=l.end()){
            tmp1.push_back(*it1);
        }
        if(it1!=l.begin()){
            it1--;
            tmp1.push_back(*it1);
        }

        auto it2=lower_bound(it(h),x);
        if(it2!=h.end()){
            tmp2.push_back(*it2);
        }
        if(it2!=h.begin()){
            it2--;
            tmp2.push_back(*it2);
        }
        for(ll y: tmp1){
            for(ll z: tmp2){
                ans=min(ans,fff(x,y,z));
            }
        }
    }
}

void solve(){
    ll nr,ng,nb;
    inp(nr,ng,nb);
    vinp(r,nr);
    vinp(g,ng);
    vinp(b,nb);
    sort(it(r));
    sort(it(g));
    sort(it(b));
    // vector<vector<vector<long long>>> possibilities;
    // vector<vector<long long>> ma;
    // ma.push_back(r);
    // ma.push_back(g);
    // ma.push_back(b);
    // f(r,g,b,possibilities,ma,0);
    ll ans=fff(r[0],g[0],b[0]);
    // for(auto & mat: possibilities){
    //     ff(mat,ans);
    // }
    ffff(ans,r, g, b);
    ffff(ans,r, b, g);
    ffff(ans,g, r, b);
    // ffff(ans,g, b, r);
    // ffff(ans,b, r, g);
    // ffff(ans,b, g, r);
    print(ans);
}
```
