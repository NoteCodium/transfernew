```python
int nthUglyNumber(int n) {
    vector<int> v(n);
    v[0]=1;
    int p2=0,p3=0,p5=0;
    for(int i=1;i<n;i++){
        int val=min({v[p2]*2,v[p3]*3,v[p5]*5});
        v[i]=val;
        //THIS IS HOW WE ARE ENSURING DUPLICATES ARE NOT THERE
        if(val==v[p2]*2) p2++;
        if(val==v[p3]*3) p3++;
        if(val==v[p5]*5) p5++;
    }
    return v[n-1];   
}
```
