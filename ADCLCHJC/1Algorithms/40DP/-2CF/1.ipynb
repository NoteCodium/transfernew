/**
 * @brief Solves Codeforces Problem 414B - Mashermoose
 * 
 * This problem involves finding the number of sequences of length k where each element 
 * divides the next element, with elements between 1 and n.
 * 
 * @see 
https://codeforces.com/contest/414/problem/B

```python
void solve(){   
    ll lastElem,len;
    inp(lastElem,len);
    //dp state 
    //dp[len][lastElement]
    //base case
    //dp[n][anyValue]=1
    ///answer 
    //dp[0][1]
    //putting a 1 at an invalid idx insures that
    //the first actual index can take any value
    //as all elements are multiple of 1
    vvl dp(len+1,vl(lastElem+1));
    for(ll i=1;i<=lastElem;i++){
        dp[len][i]=1;
    }
    for(int i=len-1;i>=0;i--){
        for(int j=1;j<=lastElem;j++){
            for(int k=j;k<=lastElem;k+=j){
                dp[i][j]=moda(dp[i][j],dp[i+1][k]);
            }
        } 
    }
    pri(dp[0][1]);
}
```
