any way we have to find the final length and not the exact value

no. of operations can be till 200000

```python
vvl dp;
void solve(){   
    ll n,m;
    inp(n,m);
    string s=to_string(n);
    ll sum=0;
    for(char c: s){
        ll d=c-'0';
        sum=moda(sum,dp[m][d]);
    }
    pri(sum);
}
 
int32_t main(){

    //precomputation starts
    dp= vvl(200000+1,vl(9+1,0));
    //0 to 9
    for(ll i=0;i<=9;i++){
        dp[0][i]=1;
    }
    //dp[i][j]
    //value of the digit j after i operations
    for(ll i=1;i<=200000;i++){
        for(ll j=0;j<=8;j++){
            dp[i][j]=dp[i-1][j+1];
        }
        dp[i][9]=moda(dp[i-1][0],dp[i-1][1]);
    }

    //precomputation ends

    int ttt;
    ttt=1;
    cin>>ttt;
    int ccc=ttt;
    while(ttt--){
        // cout<<"Case "<<(ccc-ttt)<<":"<<" ";
        solve();
    } 
    cerr<<(float)clock()/CLOCKS_PER_SEC<<" secs"<<endl;
 
}
```
