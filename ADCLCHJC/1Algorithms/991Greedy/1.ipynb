```python
class Solution {
public:
    int stoneGameVI(vector<int>& av, vector<int>& bv) {
        int n=av.size();
        vector<vector<int>> v;
        for(int i=0;i<n;i++) v.push_back({av[i]+bv[i],av[i],bv[i]});
        sort(v.begin(),v.end(),greater<vector<int>>());
        int a=0,b=0;
        for(int i=0;i<n;i++){
            if(i%2==0) a+=v[i][1];
            else b+=v[i][2];
        }
        if(a==b) return 0;
        if(a>b) return 1;
        else return -1;
    }
};
```
