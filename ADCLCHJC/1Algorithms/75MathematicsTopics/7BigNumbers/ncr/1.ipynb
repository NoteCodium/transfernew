```python
const int MAX = 1e5 + 5;
ll fact[MAX];

// Precompute factorials up to MAX
void computeFact() {
    fact[0] = 1;
    for (int i = 1; i < MAX; i++) {
        fact[i] = (fact[i - 1] * i) % mod;
    }
}

// nCr = n! / (r! * (n - r)!) % mod
ll ncr(int n, int r) {
    if (r < 0 || r > n) return 0;
    ll numerator = fact[n];
    ll denominator = (fact[r] * fact[n - r]) % mod;
    return modd(numerator, denominator);  // a / b % mod
}  
```
