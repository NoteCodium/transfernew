```python
void dfs(int node, vector<int> &vis, vector<vector<int>>& al, stack<int> &st){
    vis[node]=1;
    for(int x:al[node]){
        if(!vis[x]){
            dfs(x,vis,al,st);
        }
    }  
    st.push(node);
}

void dfs_(int node, vector<int> &vis, vector<vector<int>>& al){
    vis[node]=1;
    cout<<node<<"<-";
    for(int x:al[node]) if(!vis[x]) dfs_(x,vis,al);
}    

public:
int kosaraju(int n, vector<vector<int>>& al){
    vector<int> vis(n,0);
    stack<int> st;
    for(int i=0;i<n;i++){
        if(!vis[i]){
            dfs(i,vis,al,st);
        } 
    }

    //reversing the graph
    vector<vector<int>> alReverse(n);
    for(int i=0;i<n;i++){
        for(int x: al[i]){
            alReverse[x].push_back(i);
        }
    }
    
    vector<int> vis_(n,0);
    int ans=0;
    while(!st.empty()){
        int node=st.top();
        st.pop();
        if(!vis_[node]){
            ans++;
            dfs_(node,vis_,alReverse);
        }
        cout<<"\n";
    }
    
    return ans;
    
}
```
