insertion    
1. Inserting the new elements at its deserving position in a sorted array
2. Merge between a sorted array and a single element
3. Its optimized version is merge sort

```python
for (int i = 1; i < n; i++) {
    for (int j = i; j-1 >= 0 and v[j - 1] > v[j]; j--) {
        swap(v[j - 1], v[j]);
    }
}
```

bubble
1. Taking the big elements to the end

every round of bubbling, contains in worst case n-1 comparisons and n-1 swaps

```python
bool sorted = 0; 
while (!sorted) {
    sorted = 1;
    for (int i = 0; i <= n - 2; i++) {
        if (v[i] > v[i + 1]) {
            swap(v[i], v[i + 1]);
            sorted = 0;
        }
    }
}
```
