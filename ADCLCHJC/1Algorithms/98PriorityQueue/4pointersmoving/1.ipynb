```python
class Solution {
    public:
        vector<int> smallestRange(vector<vector<int>>& ma) {
            vector<int> ansv(2);
            int rr=ma.size();
            int ans=INT_MAX;
            int maxx=INT_MIN;
            vector<vector<int>> tmp;
            for(int i=0;i<rr;i++){
                tmp.push_back({i,0});
                maxx=max(maxx,ma[i][0]);
            } 
    
            auto cmp=[&](auto &a, auto &b){
                int x=ma[a[0]][a[1]],y=ma[b[0]][b[1]];
                return x>y;};
    
            priority_queue<vector<int> ,vector<vector<int>>,decltype(cmp)> pq(tmp.begin(),tmp.end(),cmp);
            while(!pq.empty()){
                auto p=pq.top();
                pq.pop();
                int i=p[0],j=p[1];
                int minn=ma[i][j];
                if(maxx-minn<ans){
                    ans=min(ans,maxx-minn);
                    ansv[0]=minn;
                    ansv[1]=maxx;
                }
                if(j+1<ma[i].size()){
                    pq.push({i,j+1});
                    maxx=max(maxx,ma[i][j+1]);    
                } 
                else{
                    break;
                    //from the previous question concept
                } 
            }
            return ansv;
        }
    };
```
