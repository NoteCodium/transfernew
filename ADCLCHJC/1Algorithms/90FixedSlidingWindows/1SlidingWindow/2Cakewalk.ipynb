```python
int minSwap(int v[], int n, int k) {
    int ws=0;
    for(int x: v) if(x<=k) ws++;
    if(ws==0) return 0;
    int cnt=0,maxx=0;
    for(int i=0;i<=ws-2;i++) if(v[i]<=k) cnt++;
    for(int i=ws-1;i<n;i++){
        if(v[i]<=k) cnt++;
        maxx=max(maxx,cnt);
        if(v[i-ws+1]<=k) cnt--;
    }
    return ws-maxx;
}
```
