```python
void solve(){
    int n,a,b;
    inp(n,a,b);
    vinp(v,n);
    for(int i=0;i<n;i++){
        if(i){
            v[i]+=v[i-1];
        }
    }
    
    multiset<ll, greater<ll>> se;
    //generating the first window
    for(int rightEnd=a-1;rightEnd<=b-1;rightEnd++){
        se.insert(v[rightEnd]);
    }
    ll ans=-inf;
    for(int leftEnd=0;leftEnd<=n-a;leftEnd++){
        ll leftPrefixSum=0;
        if(leftEnd){
            leftPrefixSum=v[leftEnd-1];
        }
        ans=max(ans,*se.begin()-leftPrefixSum);
        se.erase(se.find(v[leftEnd+a-1]));
        if(leftEnd+b-1+1<n){
            se.insert(v[leftEnd+b-1+1]);
        }
    }
    pri(ans);
    
    
}
```
