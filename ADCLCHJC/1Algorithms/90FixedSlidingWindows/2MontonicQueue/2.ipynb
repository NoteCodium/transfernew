We realise we can convert it to queue

```c++17
vector<long long> printFirstNegativeInteger(long long int v[],
    long long int n, long long int k) {
        vector<long long> ans;
        queue<long long> q;
        for(int i=0;i<=k-2;i++) if(v[i]<0) q.push(i);
        for(int i=k-1;i<n;i++){
        if(v[i]<0) q.push(i);
        if(q.empty()) ans.push_back(0);
        else ans.push_back(v[q.front()]);
        if(!q.empty() and q.front()==i-k+1) q.pop();
    }
    return ans;                                          
}
```
