```c++17
ll f(ll m, vl & greaterThen, vl &lessThen, ll n){
    ll taken=0;
    for(int i=0;i<n;i++){
        if(lessThen[i]>=taken and greaterThen[i]>=m-taken-1){
            taken++;
        }
    }
    return taken>=m;
}


void solve(){
    iinp(n);

    vl greaterThen(n),lessThen(n);
    for(int i=0;i<n;i++){
        cin>>greaterThen[i]>>lessThen[i];
    }

    ll l=0,h=inf,ans=-1;
    while(l<=h){
        ll m=l +(h-l)/2;
        if(f(m, greaterThen,lessThen, n)){
            ans=m;
            l=m+1;
        }
        else{
            h=m-1;
        }
    }
    pri(ans);
}
```
