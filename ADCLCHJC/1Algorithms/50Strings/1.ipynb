```python
class Solution {
public:
    bool canPermutePalindrome(string s) {
        unordered_map<int, char> m;
        for(char c: s){
            m[c]++;
        }
        int cnt=0;
        for(auto p: m){
            if(p.second%2){
                cnt++;
            }
        }
        return cnt<=1;
    }
};
```
