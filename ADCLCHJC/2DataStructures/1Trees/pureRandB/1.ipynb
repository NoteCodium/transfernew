```python
BinaryTreeNode<char> * f(string &s,int &idx, int n) {
    if(idx >= n) {
        return nullptr;
    }

    // Creating a root node having value as a character at current idx in string.
    // this cant be ? or :
    BinaryTreeNode<char> * root= new BinaryTreeNode<char> (s[idx]);
    if(idx == n-1) {
    //last char can not be ? or :
        return root;
    }
    idx++;

    if(s[idx] == '?') {
        idx++; 
        root->left = f(s, idx,n);

        idx++; 
        root->right = f(s, idx,n);
    }
    return root;
}

BinaryTreeNode<char> * toBinaryTree(string &s) {
    int n=s.size();
    int idx = 0;
    //this is necessarly non ? or :
    return f(s, idx,n);
}
```
