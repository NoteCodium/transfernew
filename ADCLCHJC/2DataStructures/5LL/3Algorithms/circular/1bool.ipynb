1. Detect a loop in a linked list.

```c++17
class Solution {
public:
    bool hasCycle(ListNode *head) {
    ListNode* s=head;
    ListNode* f=head;

    while(f and f->next){
        s=s->next;
        f=f->next->next;
        if(f==s) return 1;
        //in linear they will never meet
    }  
    return 0;
    }
};
```
