For every k jumps of the f pointer, make one jump of the s pointer.

```python
Node* s=head;
Node* f=head;
while(1){
    int cnt=k;
    while(cnt-- and f){
        f=f->next;
    }
    if(!f) break;
    s=s->next;
}
return s->data;
```
