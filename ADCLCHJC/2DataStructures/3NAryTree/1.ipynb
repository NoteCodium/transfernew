```python
/*
// Definition for a Node.
class Node {
public:
    int val;
    vector<Node*> children;

    Node() {}

    Node(int _val) {
        val = _val;
    }

    Node(int _val, vector<Node*> _children) {
        val = _val;
        children = _children;
    }
};
*/

class Solution {
    public:
        vector<vector<int>> levelOrder(Node* root) {
            queue<Node*> q;
            vector<vector<int>> ans;
            if(!root) return ans;
            q.push(root);
            while(q.size()){
                int sz=q.size();
                vector<int> tmp;
                while(sz--){
                    Node* n=q.front();
                    q.pop();
                    tmp.push_back(n->val);
                    for(auto node: n->children){
                        q.push(node);
                    }
                }
                ans.push_back(tmp);
            }
            return ans;
        }
    };
```
