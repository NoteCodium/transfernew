```c++11
void solve(){
    iinp(n);
    vinp(v,n);
    map<long long, long long> m;
    for(ll x:v){
      m[x]++;  
    }
    if(m.size()!=1 and m.size()!=2){
      pri("No");
      return;
    }
    if(m.size()==1){
      pri("Yes");
      return;
    }
    auto it=m.begin();
    ll one=it->second;
    it++;
    ll two=it->second;
    if(abs(one-two)<=1){
      pri("Yes");
      return;
    }
    pri("No");
}
```
